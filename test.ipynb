```python

import pandas as pd
import matplotlib.pyplot as plt
categories=["a","b","c","d"]
values=[5,10,15,20]

# creating vertical bar graph

plt.bar(categories, values,color='red')
plt.title("Vertical Bar graph")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simple data
Data = {
    'days': ["sunday", "monday", "Tuesday", "wednesday", "thursday", "friday", "saturday"],
    'sales': [100, 150, 200, 250, 300, 350, 400],
    'return': [2, 4, 6, 8, 9, 0, 1],
    'loss': [5, 7, 9, 10, 2, 4, 7]
}

df = pd.DataFrame(Data)
```

```python
plt.figure(figsize=(3,5))
plt.plot(df['days'], df['sales'], label='sales')
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Over the Week")
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.legend()
plt.tight_layout()       # Adjusts layout to prevent label cutoff
plt.show()
```

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=100, n_features=2, centers=3, random_state=42)

# Plot the generated data
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Generated Blobs Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
